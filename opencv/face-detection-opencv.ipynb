```python
import cv2
import gradio as gr
from PIL import Image

cascade_file = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"

HAAR_CASCADES = [
    (" ", "haarcascade_upperbody.xml"),
    ("   ", "haarcascade_eye_tree_eyeglasses.xml"),
    (" ", "haarcascade_eye.xml"),
    (" () ", "haarcascade_frontalcatface_extended.xml"),
    ("  ", "haarcascade_frontalcatface.xml"),
    ("  (alt  )", "haarcascade_frontalface_alt_tree.xml"),
    ("  (alt  1)", "haarcascade_frontalface_alt.xml"),
    ("  (alt  2)", "haarcascade_frontalface_alt2.xml"),
    ("   ()", "haarcascade_frontalface_default.xml"),
    (" ", "haarcascade_fullbody.xml"),
    ("   (  )", "haarcascade_lefteye_2splits.xml"),
    ("    (16)", "haarcascade_license_plate_rus_16stages.xml"),
    (" ", "haarcascade_lowerbody.xml"),
    (" () ", "haarcascade_profileface.xml"),
    ("   (  )", "haarcascade_righteye_2splits.xml"),
    ("  ", "haarcascade_russian_plate_number.xml"),
    ("  ", "haarcascade_smile.xml")
]
face_cascade = cv2.CascadeClassifier(cascade_file)

image_array = cv2.imread("test.jpg")

def detect_face(image_array, scale_factor, min_neighbors, min_size) :
    image_array = image_array.copy()

    bounding_boxes = face_cascade.detectMultiScale(
        image=image_array,
        scaleFactor=scale_factor,
        minNeighbors=min_neighbors,
        minSize=(min_size, min_size)
    )

    for bounding_box in bounding_boxes :
        x, y, w, h = bounding_box
        print(x, y, w, h)
        cv2.rectangle(image_array, (x, y), (x + w, y + h), (0, 255, 0), 3)

    return image_array

with gr.Blocks() as demo :

    DEFAULT_SCALE_FACTOR = 1.1
    DEFAULT_MIN_NEIGHBORS = 6
    DEFAULT_MIN_SIZE = 50

    scale_factor = gr.State(DEFAULT_SCALE_FACTOR)
    min_neighbors = gr.State(DEFAULT_MIN_NEIGHBORS)
    min_size = gr.State(DEFAULT_MIN_SIZE)

    def stream_webcam(image_array, scale_factor, min_neighbors, min_size) :
        result_image = detect_face(image_array, scale_factor, min_neighbors, min_size)
        return result_image
    
    def change_cascade(value) :
        global face_cascade
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + value)
        return value

    def click_apply(scale_factor_text, min_neighbors_text, min_size_text) :
        print(scale_factor_text, min_neighbors_text, min_size_text)
        return float(scale_factor_text), int(min_neighbors_text), int(min_size_text)
    
    gr.Markdown("# OpenCV  Haar   ")
    cascade_dropdown = gr.Dropdown(label=" ", choices=HAAR_CASCADES, value="haarcascade_frontalface_default.xml")

    with gr.Row() :
        scale_factor_textbox = gr.Text(label="scaleFactor", value=DEFAULT_SCALE_FACTOR, interactive=True)
        min_neighbors_textbox = gr.Text(label="minNeighbors", value=DEFAULT_MIN_NEIGHBORS, interactive=True)
        min_size_textbox = gr.Text(label="minSize", value=DEFAULT_MIN_SIZE, interactive=True)

    apply_button = gr.Button(" ")

    with gr.Row() :
        stream_image = gr.Image(label=" ", sources="webcam", streaming=True, mirror_webcam=False, width=640, height=480)
        result_image = gr.Image(label=" ", interactive=False, width=640, height=480)

    stream_image.stream(stream_webcam, inputs=[stream_image, scale_factor, min_neighbors, min_size], outputs=[result_image])
    cascade_dropdown.change(change_cascade, inputs=[cascade_dropdown], outputs=[cascade_dropdown])
    apply_button.click(click_apply, inputs=[scale_factor_textbox, min_neighbors_textbox, min_size_textbox], outputs=[scale_factor, min_neighbors, min_size])

demo.launch()
```
