```python
import gradio as gr

def request_stt(audio_pth) :
    
    # 1. Endpoint
    endpoint = stt_kr_endpoint

    # 2. METHOD
    
    # 3. Headers
    headers = {
        "Content-Type" : "audio/wav",
        "Ocp-Apim-Subscription-Key" : api_key
    }

    # 4. BODY : audio file
    with open(audio_pth, "rb") as audio_file :
        audio_data = audio_file.read()

    response = requests.post(endpoint, headers=headers, data=audio_data)
    response_json = response.json()
    text = response_json.get('DisplayText')

    if response.status_code != 200 :
        print("Error : ", response.status_code, response.text)
        return None

    return text

with gr.Blocks() as demo :

    def change_audio(audio_pth) :
        return audio_pth
    
    def click_send(audio_pth) :

        if audio_pth is None :
            return "마이크로 녹음해주세요"
        
        text = request_stt(audio_pth)

        return text
    
    with gr.Row() :
        input_mic = gr.Audio(sources="microphone", type="filepath", label="Input Microphone")
        input_file = gr.Audio(sources="upload", type="filepath", label="Input File")
        file_path_text = gr.Textbox(label="Transcription Output", lines=3)

    send_button = gr.Button("전송")
    output_text = gr.Textbox(label="결과 텍스트")

    input_mic.change(change_audio, inputs=[input_mic], outputs=[file_path_text])
    send_button.click(click_send, inputs=[input_mic], outputs=[output_text])

demo.launch()
```
